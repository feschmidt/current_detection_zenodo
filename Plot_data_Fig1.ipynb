```python
plt.plot(myline['Frequency (Hz)'],myline['S21dB (dB)'])
plt.show()
plt.close()

fitwidth=None,
trimwidth=5, 
ftype='A', 
doplots=False, 
margin=51
```

```python
freqs = myline['Frequency (Hz)'].values
S11complex = myline['S21re ()'] + 1j * myline['S21im ()']
S11complex = S11complex.values
params, _, _, _ = stlabutils.S11fit(freqs, S11complex, ftype='A', doplots=True)
```

```python
fitresbg = stlabutils.utils.S11fit.S11func(freqs, params, ftype='A')  # fit with background removed
fitresnobg = stlabutils.utils.S11fit.S11theo(freqs, params, ftype='A')  # fit with background removed
bgonly = -stlabutils.utils.S11fit.S11back(freqs, params)  # background only
```

```python
x,y = freqs/1e9,S11complex
xtheo,ytheo=x,fitresbg

plt.plot(x,np.real(y),'o',c='C0',markerfacecolor='none',label='Re(S$_{11}$)')
plt.plot(xtheo,np.real(ytheo),c='k')
plt.plot(x,np.imag(y),'o',c='C1',markerfacecolor='none',label='Im(S$_{11}$)')
plt.plot(xtheo,np.imag(ytheo),c='k')
plt.xlabel('Frequency (GHz)')
plt.ylabel(r'S$_{11}$ ()')
plt.xlim(7.432,7.444)
plt.legend(loc='lower right')
plt.show()
plt.close()

xraw,yraw,xraw_th,yraw_th=x,y,xtheo,ytheo
```

```python
x,y = freqs/1e9,(S11complex/bgonly+1)*np.exp(-1j*params['theta'].value)-1  # divide off the background
xtheo,ytheo = x,(fitresbg/bgonly+1)*np.exp(-1j*params['theta'].value)-1  # divide off the background

plt.plot(x,np.real(y),'o',c='C0',markerfacecolor='none',label='Re(S$_{11}$)')
plt.plot(xtheo,np.real(ytheo),c='k')
plt.plot(x,np.imag(y),'o',c='C1',markerfacecolor='none',label='Im(S$_{11}$)')
plt.plot(xtheo,np.imag(ytheo),c='k')
plt.xlabel('Frequency (GHz)')
plt.ylabel(r'S$_{11}$ ()')
plt.xlim(7.432,7.444)
plt.legend(loc='lower right')
plt.show()
plt.close()
```

```python
panel_c = {'xmeas':x,'ymeas':y,'ymeaslabel':'data','f0':params['f0'].value/1e9,
          'xtheo':xtheo,'ytheo':ytheo,'ytheolabel':'fit',
           'xraw':xraw,'yraw':yraw,'xraw_th':xraw_th,'yraw_th':yraw_th,
          'xlabel':'Frequency (GHz)','ylabel':r'|S$_{11}$| (normalized)'}
pickle.dump(panel_c,open('data_final/fig1_panel_c.pkl','wb'))
```

## quantifying

```python
tuning = pickle.load(open('data_final/fig2_panel_f0fit.pkl','rb'))
params=tuning['params']
ii = np.linspace(0,8e-6,1001)
myf0 = f0model(ii,fr=params['fr'],Lr=params['Lr'],Ic=params['Ic'])
```

```python
xx = pkldict['x1']
yy = pkldict['y1']
zz = pkldict['z1']
print('Bias point =',zz,'uA')
plt.plot(xx,yy)
vector = abs(yy)
indexes, _ = scipy.signal.find_peaks(vector,height=7,distance=15)
plt.plot(xx[indexes],yy[indexes],'o')

freqs = xx[indexes]
deltaF = np.diff(freqs)*1e9/1e6
deltaF # in MHz

myG1 = np.gradient(myf0/1e6,ii/1e-6) # MHz/uA

dfdI = interp1d(ii,myG1)(zz*1e-6)
print('at bias point: df/dI = ',dfdI,'MHz/uA')

Inoise = deltaF/dfdI
print('at bias point: Inoise =',Inoise,'uA')
```

```python
xx = pkldict['x2']
yy = pkldict['y2']
zz = pkldict['z2']
print('Bias point =',zz,'uA')
plt.plot(xx,yy)
vector = abs(yy)
indexes, _ = scipy.signal.find_peaks(vector,height=5.5,distance=20)
plt.plot(xx[indexes],yy[indexes],'o')
plt.grid()

freqs = xx[indexes]
deltaF = np.diff(freqs)*1e9/1e6
deltaF # in MHz

myG1 = np.gradient(myf0/1e6,ii/1e-6) # MHz/uA

dfdI = interp1d(ii,myG1)(zz*1e-6)
print('at bias point: df/dI = ',dfdI,'MHz/uA')

Inoise = deltaF/dfdI
print('at bias point: Inoise =',Inoise,'uA')
```

```python
xx = pkldict['x3']
yy = pkldict['y3']
zz = pkldict['z3']
print('Bias point =',zz,'uA')
plt.plot(xx,yy)
vector = abs(yy)
indexes, _ = scipy.signal.find_peaks(vector,height=4,distance=50)
plt.plot(xx[indexes],yy[indexes],'o')
plt.grid()

freqs = xx[indexes]
deltaF = np.diff(freqs)*1e9/1e6
deltaF # in MHz

myG1 = np.gradient(myf0/1e6,ii/1e-6) # MHz/uA

dfdI = interp1d(ii,myG1)(zz*1e-6)
print('at bias point: df/dI = ',dfdI,'MHz/uA')

Inoise = deltaF/dfdI
print('at bias point: Inoise =',Inoise,'uA')
```
