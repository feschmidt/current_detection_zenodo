```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
myamp = pickle.load(open("data_plots/pos2/F33_pwr_f0+1f1_amplitude.pkl","rb"))
myfreq = pickle.load(open("data_plots/pos2/F33_pwr_f0+1f1_frequency.pkl","rb"))
```

```python
Za = myamp.values
xa = myamp.axes[1].values
ya = myamp.axes[0].values
Xa,Ya = np.meshgrid(xa,ya)

Zf = myfreq.values
xf = myfreq.axes[1].values
yf = myfreq.axes[0].values
Xf,Yf = np.meshgrid(xf,yf)
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(12,4))
plt.sca(ax1)
plt.title("Amplitude")
plt.pcolormesh(Xa,Ya,Za)
plt.colorbar()
plt.sca(ax2)
plt.title("Frequency")
plt.pcolormesh(Xf,Yf,Zf)
plt.colorbar()
```

```python
i=70
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(12,4))
plt.sca(ax1)
plt.plot(Ya[:,i],Za[:,i],'.') # fluctuations relative to minimum frequency value
plt.title(Xa[0,i])
plt.sca(ax2)
plt.plot(Yf[:,i],Zf[:,i]-np.nanmin(Zf[:,i]),'.') # fluctuations relative to minimum frequency value
plt.title(Xf[0,i])
```

```python
i=30
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(12,4))
plt.sca(ax1)
plt.plot(Xa[i,:],Za[i,:],'.')
# [plt.plot(Xa[j,:],Za[j,:],'.') for j in range(Z.shape[0])]
plt.title(Ya[i,0])
plt.grid()
plt.sca(ax2)
plt.plot(Xf[i,:],Zf[i,:],'.')
# [plt.plot(Xf[j,:],Zf[j,:],'.') for j in range(Z.shape[0])]
plt.title(Yf[i,0])
plt.grid()
```
